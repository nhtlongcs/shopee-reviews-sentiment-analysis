```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
rawdf = pd.read_csv('../data/v3_sentences_pseudo.csv')
```

```python

def plot_uncertainty(df):
    df_neg = df[df['lbl'] == 0]
    df_pos = df[df['lbl'] == 1]
    plt.figure(figsize=(20, 10))
    plt.hist(df_neg['prob_positve'], bins=100, alpha=0.5, label='neg')
    plt.hist(df_pos['prob_positve'], bins=100, alpha=0.5, label='pos')

    df_neg_uncertain_threshold = df_neg['prob_positve'].quantile(0.5)
    df_pos_uncertain_threshold = df_pos['prob_positve'].quantile(0.5)
    plt.axvline(df_neg_uncertain_threshold, color='r', linestyle='dashed', linewidth=1)
    plt.axvline(df_pos_uncertain_threshold, color='r', linestyle='dashed', linewidth=1)
    plt.legend()
    plt.show()
    print(df_neg_uncertain_threshold, df_pos_uncertain_threshold)
```

```python
plot_uncertainty(rawdf)
```
